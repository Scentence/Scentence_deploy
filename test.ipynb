```python
from backend.agent.database import lookup_note_by_vector

while True:
    user_input = input("검색할 내용을 입력하세요 (종료하려면 'q' 입력): ").strip()
    
    if user_input.lower() == 'q':
        print("프로그램을 종료합니다.")
        break
        
    # 벡터 검색 실행
    try:
        result = lookup_note_by_vector(user_input)
        print(f"검색 결과: {result}")
    except Exception as e:
        print(f"오류 발생: {e}")
```
